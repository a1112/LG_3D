```python
import numpy as np
import open3d as o3d
from PIL import Image
```

```python
np_data = np.load("5.npy")
```

```python
np_data.mean()
```

```python
np_data=np_data.astype(np.float16)

np_data.mean()
```

```python
np_data
```

```python
point_cloud = open3d.geometry.PointCloud()
point_cloud.points = open3d.utility.Vector3dVector(np_data)
open3d.visualization.draw_geometries([point_cloud])
```

```python
points = []

# 遍历每个像素，将其转换为 3D 点
for y in range(np_data.shape[0]):
    for x in range(np_data.shape[1]):
        if np_data[y, x]<1000 or np_data[y, x]>60000.0:
            continue
        z = np_data[y, x]   # 将灰度值归一化到 [0, 1] 范围
        points.append([x, y, z])
```

```python
points
```

```python
points = np.array(points)

# 创建 Vector3dVector
vector3d = o3d.utility.Vector3dVector(points)
```

```python
points.mean()
```

```python
point_cloud = o3d.geometry.PointCloud()
point_cloud.points = vector3d
o3d.io.write_point_cloud("output.ply", point_cloud)
o3d.visualization.draw_geometries([point_cloud])
```

```python
points = []
colors = []
np_data
# 遍历每个像素，将其转换为 3D 点，并根据深度着色
for y in range(np_data.shape[0]):
    for x in range(np_data.shape[1]):
        z = (np_data[y, x]-54000.0)*0.003  # 将灰度值归一化到 [0, 1] 范围
        points.append([x, y, z])
        colors.append([z, 0, 1 - z])  # 根据深度设置颜色，例如蓝色到红色渐变

# 将点云列表转换为 NumPy 数组
points = np.array(points)
colors = np.array(colors)
point_cloud = o3d.geometry.PointCloud()
point_cloud.points = o3d.utility.Vector3dVector(points)
point_cloud.colors = o3d.utility.Vector3dVector(colors)

# 渲染并保存图像
vis = o3d.visualization.Visualizer()
vis.create_window(visible=False)
vis.add_geometry(point_cloud)
vis.update_geometry(point_cloud)
vis.poll_events()
vis.update_renderer()

# 保存渲染图像
image_path = "rendered_point_cloud.png"
vis.capture_screen_image(image_path)
vis.destroy_window()
```
